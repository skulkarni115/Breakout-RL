```python
import gym
from algorithms import *
import matplotlib.pyplot as plt
from IPython import display
import gymnasium
import pandas as pd
```

```python
env = gymnasium.make("ALE/MarioBros-v5")
```

# Expected SARSA

```python
#expected SARSA
num_steps = 150
gamma = 1
epsilon = 0.1
step_size = 0.7
steps,rewards = exp_sarsa(env,num_steps,gamma,epsilon,step_size)
```

```python
ratio = np.divide(np.array(rewards),np.array(steps))
```

```python
eps = []
for x in range(num_steps+1):
    eps.append(x)
trials2 = np.cumsum([0]+steps)
print(len(trials2))
dict_ = {'steps': trials2, 'episodes': eps}
df = pd.DataFrame(dict_)
df.plot(x = 'steps', y = 'episodes', title = 'EXP SARSA', ylabel = 'Episodes')
```

```python
eps = []
for x in range(num_steps):
    eps.append(x)
trials2 = np.cumsum(ratio)
dict_ = {'Cumulative rewards': trials2, 'episodes': eps}
df = pd.DataFrame(dict_)
df.plot(x = 'Cumulative rewards', y = 'episodes', title = 'Exp SARSA', ylabel = 'Episodes')
```

# Overlay Graphs

```python
#all algs
num_steps = 500
gamma = 1
epsilon = 0.1
step_size = 0.7

exp_steps,exp_rewards = exp_sarsa(env,num_steps,gamma,epsilon,step_size)
steps,rewards = sarsa(env, num_steps, gamma, epsilon, step_size)
nsteps,nrewards = nstep_sarsa(env,num_steps,gamma,epsilon,step_size)
qsteps,qrewards = q_learning(env,num_steps,gamma,epsilon,step_size)
```

```python
#get cumulative rewards
cuml_exp_rewards = np.cumsum(exp_rewards)
cuml_rewards = np.cumsum(rewards)
cuml_nrewards = np.cumsum(nrewards)
cuml_qrewards = np.cumsum(qrewards)
```

```python
eps = []
for x in range(num_steps):
    eps.append(x)
```

```python
plt.plot(eps, cuml_exp_rewards, label='Exp. SARSA')
plt.plot(eps, cuml_rewards, label='SARSA')
plt.plot(eps, cuml_nrewards, label='N-step SARSA')
plt.plot(eps, cuml_qrewards, label='Q-learning')

# Adding labels and title to the plot
plt.xlabel('Episodes')
plt.ylabel('Cumulative Rewards')
plt.title('Comparison of RL Algorithms')
plt.legend()  # Adding legend to distinguish the lines

# Display the plot
plt.show()
```

```python
#calculate the ratios and get cumulative ratio
#get cumulative rewards
exp_rewards_rat = np.cumsum(np.divide(np.array(exp_rewards),np.array(exp_steps)))
rewards_rat = np.cumsum(np.divide(np.array(rewards),np.array(steps)))
nrewards_rat = np.cumsum(np.divide(np.array(nrewards),np.array(nsteps)))
qrewards_rat = np.cumsum(np.divide(np.array(qrewards),np.array(qsteps)))
```

```python
plt.plot(eps, exp_rewards_rat, label='Exp. SARSA')
plt.plot(eps, rewards_rat, label='SARSA')
plt.plot(eps, nrewards_rat, label='N-step SARSA')
plt.plot(eps, qrewards_rat, label='Q-learning')

# Adding labels and title to the plot
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward Ratio (Rewards/Steps)')
plt.title('Comparison of RL Algorithms')
plt.legend()  # Adding legend to distinguish the lines

# Display the plot
plt.show()
```

```python
#get cumulative steps
cuml_exp_steps = np.cumsum(exp_steps)
cuml_steps = np.cumsum(steps)
cuml_nsteps = np.cumsum(nsteps)
cuml_qsteps = np.cumsum(qsteps)
```

```python
plt.plot(eps, cuml_exp_steps, label='Exp. SARSA')
plt.plot(eps, cuml_steps, label='SARSA')
plt.plot(eps, cuml_nsteps, label='N-step SARSA')
plt.plot(eps, cuml_qsteps, label='Q-learning')

# Adding labels and title to the plot
plt.xlabel('Episodes')
plt.ylabel('Cumulative Steps')
plt.title('Comparison of RL Algorithms')
plt.legend()  # Adding legend to distinguish the lines

# Display the plot
plt.show()
```
